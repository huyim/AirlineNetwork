```python
import airportsdata
import numpy as np
```

```python
import networkx as nx
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt

m = Basemap(resolution='i',
        suppress_ticks=True)

m.drawcoastlines()

m.drawcountries()

plt.figsize(200)

plt.figure(figsize=(20, 20), dpi=80)

plt.show()
```

```python
g = nx.read_gml("flightlist_20191201_20191231.gml", label="label")
g.remove_edges_from(nx.selfloop_edges(g))
```

```python
airports = airportsdata.load()
```

```python
airports['ZUGY']
```

```python
def draw_flights_on_map(airline_network):
    plt.figure(figsize=(20, 20))
    pos = {}
    for i in g:
        longtitude = airports[i]['lon']
        latitude = airports[i]['lat']
        map_long, map_lat = m(longtitude,latitude)
        pos[i] = (longtitude, latitude)
    n_flights = list(np.fromiter(nx.get_edge_attributes(g, 'value').values(),dtype=int)/100)
    # edges,n_flights = zip(*nx.get_edge_attributes(g,'value').items())
    nx.draw_networkx_nodes(airline_network,pos,
                           nodelist=g.nodes(),
                           node_size=5,
                           node_color='red',
                           alpha=0.5)
    nx.draw_networkx_edges(airline_network,pos,
                           edgelist = g.edges(),
                           edge_color=n_flights,
                           alpha=0.1)
    m.drawcoastlines()

    m.drawcountries()
draw_flights_on_map(g)
```
